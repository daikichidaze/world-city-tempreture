```python
df = pd.read_csv('../data/IMF/imf-dm-export-20231104 - NGDP_RPCH.csv', skipfooter=2)
```

```python
df.replace('no data', None, inplace=True)
```

```python
df.columns = ['Country'] + df.columns[1:].to_list()
```

```python
df.set_index(df.columns[0], inplace=True)
```

```python
for c in df.columns:
    try: 
        df[c] = pd.to_numeric(df[c])
    except ValueError as e:
        raise(e)
```

```python
plt.clf()
df[(df.index=='Japan')|(df.index=='Thailand')].T.plot()
plt.axhline(0, color='black', linewidth=0.8)
plt.show()
```

```python
asean_countries = [
    "Brunei",
    "Cambodia",
    "Indonesia",
    "Lao P.D.R.",
    "Malaysia",
    "Myanmar",
    "Philippines",
    "Singapore",
    "Thailand",
    "Vietnam"
]
```

```python
asean_data = df[df.index.isin(asean_countries)].T

fig, ax = plt.subplots(figsize=(10, 8))
asean_data.plot(ax=ax, colormap='tab20')
ax.axhline(0, color='gray', linewidth=0.8)
plt.show()
```

```python
country_region = CountryRegionDict()
df['Region'] = df.index.map(country_region.get_region)
```

```python
df_asia_europe = df[df['Region'].isin(['Asia','Europe'])][['Region']+[str(x) for x in range(2019,2029)]].dropna()
```

```python
df_asia_europe = df_asia_europe.sort_values('2028', ascending=False)
df_asia_europe.to_csv('../data/gdp_growth/asia_europe.csv')
```

```python
plt.clf()
df[(df.index=='Estonia')|(df.index=='Thailand')][[str(x) for x in range(2000,2029)]].T.plot()
plt.axhline(0, color='black', linewidth=0.8)
plt.show()
```

```python
chunk_size = 8
chunk_lists = [df_asia_europe.index[i:i + chunk_size] for i in range(0, len(df_asia_europe), chunk_size)]
```

```python
for chunk in chunk_lists:
    df_tmp = df_asia_europe[df_asia_europe.index.isin(chunk)][[str(x) for x in range(2019,2029)]].T

    plt.clf()
    fig, ax = plt.subplots(figsize=(8, 6))
    df_tmp.plot(ax=ax, colormap='tab20', marker='o')
    ax.axhline(0, color='gray', linewidth=0.8)
    plt.title(f'{df_tmp.loc["2028"].max()} - {df_tmp.loc["2028"].min()}')
    plt.show()
```
