```python
# Compare english name
df_imf = pd.read_csv('../data/gdp_growth/asia_europe.csv')
```

```python
df_imf.head(10)
```

```python
cnc = CountryNameConverter()
```

```python
country_eng_list = cnc.get_english_name_list()
```

```python
df_imf['InEngList'] = df_imf['Country'].isin(country_eng_list)
```

```python
df_imf[~df_imf['InEngList']]
```

```python
df_tempreture = pd.read_csv('../data/month_tempreture_mean.csv')
```

```python
df_tempreture[df_tempreture['mean'].isna()]
```

```python
df_tempreture['country_name_eng'] = df_tempreture['country_name'].map(cnc.get_english_name)

df_tempreture_mean = df_tempreture.groupby('country_name_eng').mean()['mean	mean_sd	max	max_sd	min	min_sd'.split('\t')]
df_tempreture_mean
```

```python
df_imf.set_index('Country', inplace=True)
df_imf
```

```python
df_merged = pd.concat([df_tempreture_mean, df_imf], axis=1)
df_merged.drop('InEngList', axis=1, inplace=True)
```

```python
df_viz = df_merged[(df_merged['mean'].notna())&(df_merged['2028'].notna())]
df_viz.index.name = 'Country'
df_viz.to_csv('viz.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
df_viz[df_viz['Region']=='Asia'].plot.scatter('2028', 'mean', c='blue', label='Asia', ax=ax)
df_viz[df_viz['Region']=='Europe'].plot.scatter('2028', 'mean', c='red', label='Europe', ax=ax)
plt.show()
```
