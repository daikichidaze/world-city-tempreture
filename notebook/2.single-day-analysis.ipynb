```python
def one_day_data(input_path: str) -> pd.DataFrame:
    df = pd.read_csv(input_path)
    df.columns = col_names
    df['date'] = datetime.strptime(input_path[-14:-4],'%Y-%m-%d')
    return df

df = pd.concat([one_day_data(p) for p in glob('../data/raw/*csv')], axis=0).reset_index(drop=True)
    
```

```python
# no duplicated poi_id
for poi_id in df['poi_id'].drop_duplicates().tolist():
    gb_name = df[df['poi_id']==poi_id].groupby(['poi_name', 'country_name']).count()
    if len(gb_name) > 1:
        raise(f'poi_id = {poi_id}, length = {len(gb_name)}')
```

```python
df_count_days = df.groupby(['poi_id', 'poi_name', 'country_name']).count()['date'].reset_index()
```

```python
df_count_days['date'].value_counts()
```

```python
df_count_days[df_count_days['date']==364].to_csv('../data/tmp.csv')
```

```python
year_set = set(datetime(2021,1,1) + timedelta(days=i) for i in range(365))


target_pois = df_count_days[(360 <=  df_count_days['date']) & (df_count_days['date'] < 365)]

result = {}

for poi_id in target_pois['poi_id']:
    df_id = df[df['poi_id']==poi_id]
    loss_days = year_set - set(df_id['date'])
    result[poi_id] = loss_days
```
